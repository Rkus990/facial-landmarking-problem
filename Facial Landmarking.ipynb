We observe that representative parts of the face (e.g., eyes) are represented in the two systems by different number of points, however, essentially the objectives of such points is only to increase the density in that local region only.

For example, 

- Jaw is represented by points with indices 0,1,...16 in 300W (68-point) system.
- Jaw is represented by points with indices 0,1,2....,32 in WFLW (98-point) system

Similarly, 
- Left-eyebrow = 17,18,....21 (300W)
- Left-eyebrow = 22,23,....26 (WFLW)

;
- Right-eyebrow = 22,23,...26 (300W)
- Right-eyebrow = 27,28,...35 (WFLW)

;
- Nose = 27,28,...35 (300W)
- Nose = 51,52,...59 (WFLW)

;
- Left eye = 36,37,...41 (300W)
- Left eye = 60,61,...67,96 (WFLW)

;
- Right eye = 42,43,....47 (300W)
- Right eye = 68,69,....75,97 (WFLW)

;
- Mouth = 48,49,....67 (300W)
- Mouth = 76,77,....95 (WFLW)

We further make the references to the source github repository and we thank the author and requestor on the issue [#17](https://github.com/wywu/LAB/issues/17) on Github.

We make changes in their proposition stating that certain features such as "eyes" (both), "eyebrows" (both) , "left_eye_poly", "right_eye_poly" and "eyes_and_eyebrows" seem redundant and may be omitted.

**Question 2 
One set of WFLW-style landmarks has 196 degrees of freedom (2 for each of the 98 keypoints).**
- **Create a function that normalizes WFLW-style landmarks by “removing” degrees of freedom for
Rotation in the plane of the image
Translation in the plane of the image
Scaling
(i.e. the input to your function should be a set of WFLW-style landmarks, and the output should be a transformed set of WFLW-style landmarks. The output should be approximately invariant under rotation, translation, and scaling of the input).**



- **Use your function to create a WFLW_normalized_landmarks_test.csv by applying it to each row of the WFLW_landmarks_test.csv file.**


This means we will use:
- df to refer to WFLW-train
- df_test -> WFLW-test
- df_test_t -> WFLW-test-translation-invariant dataset
- df_test_tr -> WFLW-test-translation-and-rotation-invariant-dataset
- **df_test_trs** -> WFLW-test-translation-rotation-and-scaling-invariant-dataset -> this is normalized or the final **WFLW_normalized_landmarks_test.csv** file

For translation, idea is to find the centroid of all the points and then calculate the difference of all the annotation points from this centroid, thus essentially making the centroid of each image as the origin. Thus, this image is easily translatable without variation, as long as it is done with respect to the centroid (our assumption and frame of reference). 

```python
def translate_invariance(df_sample):
    df_sample_x = df_sample.filter(regex='^x') #df_sample.iloc[:, 1::2]
    df_sample_y = df_sample.filter(regex='^y') #df_sample.iloc[:, 2::2]
    
    # calculating centroid
    df_sample_x_mean = df_sample_x.mean(axis=1)
    df_sample_y_mean = df_sample_y.mean(axis=1)
    
    
    # Subtracting from all points the centroid
    df_sample_x = df_sample_x.sub(df_sample_x.mean(axis=1), axis=0)
    df_sample_y = df_sample_y.sub(df_sample_y.mean(axis=1), axis=0)
    
    cols_x = list(df_sample_x.columns)
    cols_y = list(df_sample_y.columns)
    
    ans_sample = df_sample.copy()
    
    ans_sample[cols_x] = df_sample_x[cols_x]
    
    ans_sample[cols_y] = df_sample_y[cols_y]
    
    return ans_sample
```

Here, 
- the idea is to first find the rotation angle with reference to a point that it makes with x-axis (or the slope) about which we rotate the image(in this case we use first point).
- then we rotate all the points for each of the faces by negative of this rotation angle, essentially undoing the effects of rotation in relative sense.

# Rotation_invariant function, finally!

```python
def rotation_invariance(df_sample):
    
    df_sample_x = df_sample.filter(regex='^x') #df_sample.iloc[:, 1::2]
    df_sample_y = df_sample.filter(regex='^y') #df_sample.iloc[:, 2::2]
    
    # Caching some values
    index_x = df_sample_x.index
    column_x = df_sample_x.columns
    
    index_y = df_sample_y.index
    column_y = df_sample_y.columns
    
    
    # 499 rotation angles, 1 for each face
    rotation_angle = -np.arctan2(df_sample_y.iloc[:,0], df_sample_x.iloc[:,0])
    
    # 2*2*499 matrix
    rotation_matrix = np.array([[np.cos(rotation_angle), -np.sin(rotation_angle)],
                                 [np.sin(rotation_angle), np.cos(rotation_angle)]])
    
    # 499*2*2 -> 499 faces, for each face, there is a 2x2 rotation matrix
    rotation_matrix = rotation_matrix.transpose(2,0,1) # (499,2,2)
    
    # stack 499*98*2 by stacking 2 sets of 499*98 x and y each  
    st_temp = np.stack((df_sample_x, df_sample_y), axis=-1)
    
    # stack transpose taken -> 499*2*98 for multiplication
    st_temp = st_temp.transpose(0,2,1) # (499,2,98)
    
    # 499*2*98 -> multiplying stacked matrix of 499*2*2 times 499*2*98
    after_rotation = np.matmul(rotation_matrix, st_temp)
    
    # 499*98*2 -> bring back in the original form
    after_rotation = after_rotation.transpose(0,2,1)
    
    #499*98 -> x points -> bring back in dataframe form
    df_sample_x = pd.DataFrame(data = after_rotation[:,:,0],  
                  index = index_x,  
                  columns = column_x)
    
    #499*98 -> y points -> bring back in dataframe form
    df_sample_y = pd.DataFrame(data = after_rotation[:,:,1],  
                  index = index_y,  
                  columns = column_y)
    
    #getting the original order of x and y of csv file
    ans_sample = df_sample.copy()
    
    cols_x = list(df_sample_x.columns)
    cols_y = list(df_sample_y.columns)
    
    # pasting the x and y in a dataframe in the original order        
    ans_sample[cols_x] = df_sample_x[cols_x]
    
    ans_sample[cols_y] = df_sample_y[cols_y]
    
    return ans_sample
```

## Finally, scaling function ready!

```python
def scaling_invariance(df_sample):
    df_sample_x = df_sample.filter(regex='^x') #df_sample.iloc[:, 1::2]
    df_sample_y = df_sample.filter(regex='^y') #df_sample.iloc[:, 2::2]
    
    # calculating centroid
    df_sample_x_mean = df_sample_x.mean(axis=1)
    df_sample_y_mean = df_sample_y.mean(axis=1)
    
    # reciprocal of the average distance between norm of annotated points
    x_factor = 1/np.linalg.norm(df_sample_x, axis=1).mean()
    
    y_factor = 1/np.linalg.norm(df_sample_y, axis=1).mean()
    # Multiplying each point by the scaling factor to normalize
    df_sample_x = df_sample_x*x_factor
    df_sample_y = df_sample_y*y_factor
    
    cols_x = list(df_sample_x.columns)
    cols_y = list(df_sample_y.columns)
    
    ans_sample = df_sample.copy()
    
    ans_sample[cols_x] = df_sample_x[cols_x]
    
    ans_sample[cols_y] = df_sample_y[cols_y]
    
    return ans_sample
```
